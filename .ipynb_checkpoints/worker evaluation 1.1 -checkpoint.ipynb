```python
# import data
raw_df = pd.read_csv('f835051.csv')
# take out the golden training data (skews things dramatically)
raw_df = raw_df[raw_df['_golden'] == False]
```

```python
# remove the entries for which there is no destination url 
nan_list = []

for endurl in raw_df['endurl']:
    nan_list.append(type(endurl) != type(float()))

    
reduced_raw_df = raw_df[nan_list]      
print reduced_raw_df.shape
print raw_df.shape

raw_df = reduced_raw_df
```

```python
#extract the information that i want from the dataframe object

worker_ids = []
unit_ids = []
worker_scores = []
event_id = []
event_time_stamp = []
start_time_stamp = []
trust = []


count = 0
test_size = 100

for row in raw_df.groupby('_worker_id'):
    worker_ids.append(row[0])
    unit_id.append(row[1]['_unit_id'])
    worker_scores.append(row[1]['please_make_your_relevancy_judgment'])
    event_id.append(row[1]['_id'])
    trust.append(row[1].append)
    event_time_stamp.append(row[1]['_started_at'])
    start_time_stamp.append(row[1]['_created_at'])

    
    count += 1
    if count == test_size:
        break
```

```python
# make some aggregations into the objects with the more useful information paired up and so forth
id_scores_dict = dict(zip(worker_ids, worker_scores))
id_time_dict = dict(zip(worker_ids, event_time_stamp))
```

```python
from datetime import datetime, date, time, timedelta

def inter_times(list_of_times):
    between_times = []
    counter = 0
    
    time_object_list = []
    
    #convert to timedate objects 
    for time_stamp in list_of_times:
#        print time_stamp
        as_time_object = datetime.strptime(time_stamp, "%m/%d/%Y %H:%M:%S")  
        time_object_list.append(as_time_object)
    
    # sort this object 
#    time_object_list.sort()
    
    for now_time in time_object_list:
        # convert to datetime object 
        now = now_time 

        # on the first time through, just create an dt object and start next iteration 
        if counter == 0:
            counter +=1 
            then = now
            continue 
                    
        diff = (now - then).total_seconds()
        then = now
        between_times.append(diff)            
    return between_times
```

```python
time_differences = []

for event_times in event_time_stamp:
    time_differences.append(inter_times(event_times))
    
event_df = pd.DataFrame(data = time_differences).transpose()
```

```python
def make_list_datetime(list_of_time_strings):
    date_time_list = []
    for time in list_of_time_strings:
        date_time_list.append(datetime.strptime(time, "%m/%d/%Y %H:%M:%S"))
    return date_time_list

print make_list_datetime(start_time_stamp[0])
```

```python
#maybe there's something about the start and creation time differences that matters 

def creation_start_dif(creation_list, start_list):
    creation_start_diffs = []
    
    creation_datetimes = make_list_datetime(creation_list)
    start_datetimes = make_list_datetime(start_list)
    
    # make all into datetime objects 
    
    indexer = range(0,len(creation_datetimes))

    for index in indexer:
        diff = creation_datetimes[index] - start_datetimes[index]
        creation_start_diffs.append(diff.total_seconds())
    
    return creation_start_diffs
```

```python
creation_start_diff_list = []

indexer = range(0,len(start_time_stamp))
for index in indexer:
    diff_list = creation_start_dif(start_time_stamp[index], event_time_stamp[index])
    creation_start_diff_list.append(diff_list)
```

```python
diversity_list = []
for ind_list in creation_start_diff_list:
    uniques = len(set(ind_list)) + 0.0
    diversity = uniques/len(ind_list)
    diversity_list.append(diversity)
pd.Series(data=diversity_list).hist(bins = 10)
```

```python
diversity_id_dict = dict(zip(diversity_list, worker_ids))
diversity_id_dict[min(diversity_list)]

print min(diversity_list)
##THIS IS A ROBOT!!!
#i have successfully identified another way to to spot robots 
```

```python
## how to fit a normal distribution to a histogram and do basic anomaly detection  
diversity_series = pd.Series(data = diversity_list)

mu = diversity_series.mean()
sigma = diversity_series.std()
n = len(diversity_series)

print mu, sigma

min_ci_90 = mu - 1.64*(mu*(1-mu)/n)**.5
print min_ci_90
#funny that the score comes in just below the confidence interval 
```
